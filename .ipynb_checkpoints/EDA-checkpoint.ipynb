```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```


The post has a variable number of different tags. The below plot shows the tags in different sizes based on its popularity

```python
df.Tags = df.Tags.apply(lambda x: x.split('|'))
tag_freq_dict = {}
for tags in df.Tags:
    for tag in tags:
        if tag not in tag_freq_dict:
            tag_freq_dict[tag] = 0
        else:
            tag_freq_dict[tag] += 1
```

```python
wordcloud = WordCloud(    background_color='black',
                          width=1600,
                          height=800,
                    ).generate_from_frequencies(tag_freq_dict)

fig = plt.figure(figsize=(30,20))
plt.imshow(wordcloud)
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```

Here, I'm filtering the most popular 500 tags based on their occurences. Given the dataset has 500k values, 500tags seems to be a good number to experiment with.

```python
tags_to_use = 500
tag_freq_dict_sorted = dict(sorted(tag_freq_dict.items(), key=lambda x: x[1], reverse=True))
final_tags = list(tag_freq_dict_sorted.keys())[:tags_to_use]
len(final_tags)
```

```python
final_tag_data = []
for tags in df.Tags:
    temp = []
    for tag in tags:
        if tag in final_tags:
            temp.append(tag)
    final_tag_data.append(temp)
```

```python
# df['Corpus'] = df['Corpus'].map(lambda x: x.replace(r'\D', '')('thanks'))
# df['Corpus'] = df['Corpus'].map(lambda x: x.lstrip('thank'))


df['Corpus'] = df['Corpus'].str.replace('thanks', '')
df['Corpus'] = df['Corpus'].str.replace('thank', '')



```

```python
from sklearn.preprocessing import MultiLabelBinarizer
tag_encoder = MultiLabelBinarizer()
tags_encoded = tag_encoder.fit_transform(final_tag_data)
tags_encoded.shape
```

```python
# Word2Vec
import gensim

# WORD2VEC 
W2V_SIZE = 300
W2V_WINDOW = 7
W2V_EPOCH = 32
W2V_MIN_COUNT = 10

# Collect corpus for training word embeddings
documents = [_text.split() for _text in np.array(df.Corpus)]  # I used the entire corpus for training, because it contains the entire conversation of related to a user's posts
```

```python
df.Corpus[0]
```

```python
w2v_model = gensim.models.word2vec.Word2Vec(size=W2V_SIZE, 
                                            window=W2V_WINDOW, 
                                            min_count=W2V_MIN_COUNT, 
                                            workers=8)
w2v_model.build_vocab(documents)
words = w2v_model.wv.vocab.keys()
vocab_size = len(words)
print("Vocab size", vocab_size)
# Train Word Embeddings
w2v_model.train(documents, total_examples=len(documents), epochs=W2V_EPOCH)
w2v_model.save('SO_word2vec_embeddings.bin')
```

```python
w2v_model = gensim.models.word2vec.Word2Vec.load(os.getcwd()+'\\SO_word2vec_embeddings.bin')
```

Sanity test to check on the trained word embeddings model

```python
print('Terms most similar to "numpy"')
print(w2v_model.wv.most_similar("numpy"))
print('-'*100)

print('Terms most similar to "pandas"')
print(w2v_model.wv.most_similar("pandas"))
print('-'*100)

print('Terms most similar to "javascript"')
print(w2v_model.wv.most_similar("javascript"))
print('-'*100)

print('Terms most similar to "jsp"')
print(w2v_model.wv.most_similar("jspe"))
```
